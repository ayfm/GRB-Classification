<ol>

    <ol>

    <ol>
        <li>T90, </li> 
        <li>Spectral Hardness (100-300 / 25-50 keV), </li>

    </ol>

    <ol>
        <li>T90, </li> 
        <li>T90i, </li>

        <li>Spectral Hardnesss-Luminosty, </li> 
        <li>T90i-Spectral Hardness-Luminosity. </li> 
    </ol>

</ol>

Please note that each feature space is in log10 base.
